```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

plt.rcParams.update({'font.size': 16})
```

```python
df = pd.read_excel("DataLab2Student.xlsx", header = 514, usecols = "B, I, J, L, M, O, P, R, S, U, V", 
                   names = ["Loss", "VaR-N", "ES-N", "VaR-N-EWMA", "ES-N-EWMA", "VaR-T", "ES-T", "VaR-T-EWMA",
                           "ES-T-EWMA", "VaR-BHS", "ES-BHS"])
df.head(10)
df = df.dropna()
```

```python
fig = plt.figure(figsize=(20,15))
gs = gridspec.GridSpec(3,4)

ax0 = plt.subplot(gs[0,0:2])
ax0.plot(df["Loss"],  label = "Loss")
ax0.plot(df["VaR-N"], label = "VaR")
ax0.plot(df["ES-N"], label = "ES")
ax0.set_xlim(0, 250)
ax0.set_ylim(-15,45)
ax0.set_title("Normal")
ax0.legend(loc = 'upper left')

ax1 = plt.subplot(gs[0,2:4])
ax1.plot(df["Loss"], label = "Loss")
ax1.plot(df["VaR-N-EWMA"], label = "VaR")
ax1.plot(df["ES-N-EWMA"], label = "ES")
ax1.set_xlim(0, 250)
ax1.set_ylim(-15,45)
ax1.set_title("Normal-EWMA")
ax1.legend(loc = 'upper left')

ax2 = plt.subplot(gs[1,0:2])
ax2.plot(df["Loss"], label = "Loss")
ax2.plot(df["VaR-N-EWMA"], label = "VaR")
ax2.plot(df["ES-N-EWMA"], label = "ES")
ax2.set_xlim(0, 250)
ax2.set_ylim(-15,45)
ax2.set_title("T-dist")
ax2.legend(loc = 'upper left')

ax3 = plt.subplot(gs[1,2:4])
ax3.plot(df["Loss"], label = "Loss")
ax3.plot(df["VaR-N-EWMA"], label = "VaR")
ax3.plot(df["ES-N-EWMA"], label = "ES")
ax3.set_xlim(0, 250)
ax3.set_ylim(-15,45)
ax3.set_title("T-dist-EWMA")
ax3.legend(loc = 'upper left')

ax4 = plt.subplot(gs[2,1:3])
ax4.plot(df["Loss"], label = "Loss")
ax4.plot(df["VaR-N-EWMA"], label = "VaR")
ax4.plot(df["ES-N-EWMA"], label = "ES")
ax4.set_xlim(0, 250)
ax4.set_ylim(-15,45)
ax4.set_title("BHS")
ax4.legend(loc = 'upper left')

plt.show()
```
